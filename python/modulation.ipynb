```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import scipy as sp
from scipy import signal
```

```python
Fs = 10**3
Ts = 1/Fs
t = np.arange(0, 1, Ts)
data = [random.randint(0,1) for _ in range(len(t))]
```

```python
def modulation(input_signal, modulationType, fc, t):
    if modulationType == "ASK":
        carrier = np.sin(2 * np.pi * fc * t)  
        modulated = carrier * np.array(input_signal) 
        return carrier, modulated
```

```python
fig, axs = plt.subplots(3, 1, figsize=(20, 15))  # Set the figure size
carrier, modulated = modulation(data, "ASK", 100, t)
axs[0].plot(t, data, 'b', linewidth=2, drawstyle='steps-pre')
axs[0].set_ylim([-0.5, 1.5])
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Message Signal')

axs[1].plot(t, carrier, 'r', linewidth=2)
axs[1].set_ylim([-2.5, 2.5])
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Carrier Signal')

axs[2].plot(t, modulated, 'g', linewidth=2)
axs[2].set_ylim([-2.5, 2.5])
axs[2].set_xlabel('Time (s)')
axs[2].set_ylabel('Amplitude')
axs[2].set_title('Modulated Signal')

for i in range(0,3):
    axs[i].set_xlim([0, 0.5])
    axs[i].grid()
```

```python
def binary(sym, sym_len):

  import numpy as np
  rand_n = np.random.rand(sym)
  rand_n[np.where(rand_n >= 0.5)] = 1
  rand_n[np.where(rand_n <= 0.5)] = 0

  sig = np.zeros(int(sym*sym_len))

  # generating symbols
  id1 = np.where(rand_n == 1)

  for i in id1[0]:
    temp = int(i*sym_len)
    sig[temp:temp+sym_len] = 1
  return sig
```

```python
Fs = 1000 # Samples per second
fc = 25 # Carrier frequency 100 Hz, 100 cycles/sec
T = 1 # Total simulation time in seconds
t = np.arange(0, T, 1/Fs)
x =np.sin(2*np.pi*fc*t)
Td = 0.1 # Bit duration
Nsamples = int(Td*Fs) # Samples in one bit duration
Nsym = int(np.floor(np.size(t)/Nsamples))
# Python code to generate binary stream of data
sig = binary(Nsym, Nsamples)
# ASK waveform generation
```
