```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PATH = "../dataset/"
labels = np.load(f'{PATH}/labels.npy', mmap_mode = 'r')
snrs = np.load(f'{PATH}/snrs.npy', mmap_mode = 'r')
signals = np.load(f'{PATH}/signals.npy', mmap_mode = 'r')

with open(f'{PATH}/classes.txt') as f:
    classes = f.read()

def str_to_list(line):
    line = line.replace('\n','')
    line = line.replace('classes = ','')
    line = eval(line)
    return line
    
classes = str_to_list(classes)
print(classes)
```

```python
np.unique(snrs)
```

```python
idx = 1
# Information about the signals[idx]
print(f'Signal index {idx}\n'
      f'\t- Classes: {classes[idx]}\n'
      f'\t- Modulation: {labels[idx]}\n'
      f'\t- SNR: {snrs[idx]}\n'
      f'\t- Samples: {signals[idx]}')

fig = plt.figure(figsize=(20, 6), dpi=80)
plt.plot(signals[idx][:, 0], color='green', label='I component')
plt.plot(signals[idx][:, 1], color='salmon', label='Q component')
plt.title(classes[idx])
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
```
